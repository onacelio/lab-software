```python
import pandas as pd
```

```python
vendasData = pd.read_csv(r"lista-de-vendas-2017.csv", sep=";")
lista_colunas = ['ID Produto', 'Quantidade Vendida', 'Quantidade Devolvida']
vendasData[lista_colunas]
```

```python
quantidadeDeVenda = int(input("Insira a quantidade de vendas: "))
numeroVenda = vendasData["Numero da Venda"]
quantidadeVendida = vendasData["Quantidade Vendida"]
idProduto = vendasData["ID Produto"]

for i in range(0, len(numeroVenda)):
    if quantidadeVendida[i] >= quantidadeDeVenda:
        print(f"Quantidade vendida: {quantidadeVendida[i]}. Id do produto: {idProduto[i]}")
```
